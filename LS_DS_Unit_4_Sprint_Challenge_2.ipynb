## 1. Define the following terms:

- **Neuron:** In biology, a neuron takes in an input through it's dendrites, leading to action potential, if the signal is huge enough, it is passed down its axon to the next cell body. In computer science, neurons take in an input, assign some weight to it, perform computation and activation and gives us an output
- **Input Layer:** Input is the features we give the neuron network. Typically we have one input node for each feature/column that we give our neuron
- **Hidden Layer:** This is invisible to us and is where the neuron performs computation, makes connection between features and and perform their calculation. We do not interact with hidden layers directly. The most simple neural network has only one hidden layer.
- **Output Layer:** This is what gives us the resuls of the computation from the hidden layers. We typically use activation functions to change the output to a format that makees sense for our problem. Different types of problems have different types of activation functions for the output layer
- **Activation:** In neural networks, activation functions decide on how much signal to pass onto to next layer. For a simple classification problem, we use a sigmoid function but for multi-class we might use something as a softmax function which gives each output probabilities, which add up to 1. 
- **Backpropagation:** Backpropagation is a method where the weights of the inputs are updated after completion of one cycle of going through the model. This way we can see which inputs are giving the most desirable results when we get to the output layer and in the next cycle, more weight is assigned to them to train the model.

```python
class NeuralNetwork:
    def __init__(self):
        # Set up Architecture
        self.inputs = 13
        self.hiddenNodes = 13
        self.outputNodes = 1

        #Initial weights
        self.weights1 = np.random.randn(self.inputs, self.hiddenNodes) #2x3
        self.weights2 = np.random.rand(self.hiddenNodes, self.outputNodes) #3x1

    def sigmoid(self, s):
        return 1 / (1+np.exp(-s))

    def sigmoidPrime(self, s):
        return s * (1 - s)

    def feed_forward(self, X):
        """
        Calculate the NN inference using feed forward.
        """

        #Weighted sume of inputs and hidden layer
        self.hidden_sum = np.dot(X, self.weights1)

        #Acivations of weighted sum
        self.activated_hidden = self.sigmoid(self.hidden_sum)

        # Weight sum between hidden and output
        self.output_sum = np.dot(self.activated_hidden, self.weights2)

        #Final activation of output
        self.activated_output = self.sigmoid(self.output_sum)

        return self.activated_output

    def backward(self, X, y, o):
        """
        Backward propagate through the network
        """
        self.o_error = y - o #error in output
        self.o_delta = self.o_error * self.sigmoidPrime(o) # apply derivative of sigmoid to error

        self.z2_error = self.o_delta.dot(self.weights2.T) # z2 error: how much our hidden layer weights were off
        self.z2_delta = self.z2_error*self.sigmoidPrime(self.activated_hidden)

        self.weights1 += X.T.dot(self.z2_delta) #Adjust first set (input => hidden) weights
        self.weights2 += self.activated_hidden.T.dot(self.o_delta) #adjust second set (hidden => output) weights

    def train(self, X, y):
        self.X = X
        self.y = y
        o = self.feed_forward(X)
        self.backward(X, y, o)
```
